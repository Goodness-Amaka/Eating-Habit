## Lets separate the numerical datatypes from objects

```python
numeric = EH.select_dtypes(np.number)
numeric
```

```python
objects = EH.select_dtypes('object')
objects
```

```python
# The object datatypes will now be converted to numerical category in order to pass it into our algorithm
# Labelenconder converts object values into numerics

from sklearn.preprocessing import LabelEncoder
```

```python
# instanciation

Label_encoder = LabelEncoder()
Label_encoder
```

```python
# using a variable to get all the object datatypes columns as a list because i want to use for loop iteration
# method 1

obj = objects.columns.to_list()
obj
```

```python
# you can call the object columns out directly as well by:

objects.columns
```

```python
# brought in the function that converts objects datatypes to numerics to help convert object(string)values

# format 1
# for columns in obj:
#     objects[column] = Label_encoder.fit_transform(objects[column])
 
# format2
    
for i in objects.columns:
    objects[i] = Label_encoder.fit_transform(objects[i])
    
```

```python
objects
```

```python
EH1 = numeric.merge(objects, right_index=True, left_index=True)
EH1
```

```python
# creating our x

EHx = EH1[['Age','I avoid eating when am hungry','I cut my food into small pieces',
           'I particularly avoid food with high carbonhydrate(bread,rice,potatoes etc)',
           'I take longer time than others to eat my meals', 
           'I avoid food and drink with sugar content', 'I like my stomach empty ', 'I fast to avoid eating']].values
EHx
```

```python
EHy = EH1['Class'] 
EHy
```

```python
# ERROR COPY

# now that we have created the features and labels (input(x), output(y)),
# we are going to split the data into training and testing

from sklearn.model_selection import train_test_split


# EHx_train ,EHx_test, EHy_train, EHy_teSt = train_test_split (EHx, EHy, train_size= 0.20, random_state=0)
```

```python
EHx_train ,EHx_test, EHy_train, EHy_test = train_test_split (EHx, EHy, train_size= 0.20, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
KNN_model = KNeighborsClassifier()


DTC_model = DecisionTreeClassifier()

SVC_model = SVC()
```

```python
# Training the model

SVC_model.fit(EHx_train, EHy_train)
KNN_model.fit(EHx_train, EHy_train)
DTC_model.fit(EHx_train, EHy_train)
```

```python
# Prediction


SVC_pred = SVC_model.predict(EHx_test)
KNN_pred = KNN_model.predict(EHx_test)
DTC_pred = DTC_model.predict(EHx_test)
```

```python
from sklearn.metrics import accuracy_score, r2_score, confusion_matrix
```
